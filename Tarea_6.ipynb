```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 1e-5 #desviaciación estandar
val_exacto = 1 - np.cos(1)
def f(x): #funcion sinusoidal
  return np.sin(x)

def p_n(x):#Polinomio con error
  n = (np.random.normal(0, sigma, size = x.shape))
  return np.sin(x) + n

#Metodo de simpson
def simpson(a, b, N):
  if N%2 ==1:
    N+=1
  h = (b-a)/N
  x = np.linspace(a, b, N+1)
  y = p_n(x)

  integral = h/3*(y[0]+ 2 * np.sum(y[2:N:2]) + 4 * np.sum(y[1:N:2]) + y[N])
  return integral

simpson(0 ,1 ,100)


val_N = np.arange(10, 500, 10) # valores para n
val_h= (1-0)/val_N


error = []
for N in val_N:
    val_int = simpson(0, 1, N)
    error_abs = np.abs(val_int - val_exacto)
    error.append(error_abs)



# Graficar el error
plt.figure(figsize=(8, 6))
plt.loglog(val_h, error, marker='o', linestyle='-', color='b')
plt.title("Error cometido por el método de Simpson compuesto en función de $h$")
plt.xlabel("Tamaño del subintervalo $h$")
plt.ylabel("Error absoluto")
plt.grid(True, which="both", ls="--")
plt.show()
```

En base a lo obtenido, si se puede alcanzar una presición mejor que $10^{-7}$, esto debido a que el error fue minimizado hasta estár por debajo de $10^{-7}$, pero se puede observar una gran cantidad de ruido en la gráfica presente, por lo que el resultado debido a un error de la naturaleza experimental del proceso físico correspondiente, presentara una gran variabilidad en torno al cálculo de la integral.
