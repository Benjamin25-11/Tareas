```python
import matplotlib.pyplot as plt
import numpy as np

#Posiciones en cm
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

#Tiempo (s) en intervalos regulares
arreglo_tiempo = np.linspace(0,7,8)

#Error en cm
error = 0.4

#Figura
fig = plt.figure()

#Ejes
axs1 = fig.add_subplot(1,2,1)
axs2 = fig.add_subplot(1,2,2)

#Datos
axs1.errorbar(arreglo_tiempo, datos, yerr = error, label= 'Datos con error 4 mm')
axs2.errorbar(arreglo_tiempo, datos_7, yerr = error, label='Datos con error 4 mm')

#Titulos
fig.suptitle('Grafico de datos obtenidos por alumnos')
axs1.set_title('Datos Alumno nota 1')
axs2.set_title('Datos Alumno nota 7')
axs1.set_xlabel('Tiempo')
axs1.set_ylabel('Posición')
axs2.set_xlabel('Tiempo')
axs2.set_ylabel('Posición')

#Leyenda
axs1.legend()
axs2.legend()

#Muestro la figura
plt.tight_layout() #Ajuste para que no se superpongan ambas figuras
plt.savefig('Notas.pdf')
plt.show()
```

Considerando que el aparato utilizado para medir la posición tiene un error de mas menos 4 mm y considerando la naturaleza experimental del procedimiento, es posible esperar algunos errores en las mediciones. Observando la posición a lo largo del tiempo de la primera figura, es posible notar que la curva es casi perfecta, algo que parece extraño considerando lo expresado anteriormente. Por otro lado, el segundo alumno, presenta datos que son consistentes con el tipo de error esperable de un procedimiento experimental.
A partir de esto, es posible concluir que el primer alumno ajustó los valores obtenidos en el experimento, para poder tener una curva perfecta, por lo que merece la nota minima.
