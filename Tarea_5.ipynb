```python
import numpy as np
import matplotlib.pyplot as plt


class MaxIterations(Exception):
    pass

# Definición de la función por tramos
def f(x):
    if x <= 1:
        return x + 2
    else:
        return x - 2

# Método de bisección
def biseccion(f, a, b, epsilon, N=1000000):
    pp = a
    for i in range(N):
        p = a + (b - a) / 2
        if abs(pp - p) / abs(p) < epsilon:
            return p, i + 1
        elif f(p) * f(a) > 0:
            a = p
        else:
            b = p
        pp = p

    raise MaxIterations("No se encontró el cero luego de", N, "iteraciones.")

# Puntos de la función
x = np.linspace(-10, 10, 100)
y = np.array([f(val) for val in x])

# Se grafica la función
plt.scatter(x, y, label='f(x)', color='blue', s=10)
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Gráfica de la función por tramos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-5, 10)
plt.xlim(-10, 10)
plt.show()

# Se escoge un intervalo cercano a la discontinuidad
a = 0
b = 1.5
epsilon = 1e-7

try:
    root, iterations = biseccion(f, a, b, epsilon)
    print(f'Raíz encontrada: {root} en {iterations} iteraciones.')
except MaxIterations as e:
    print(e)

```

Si se considera una función por tramos, la cual presenta una discontinuidad, si se aplica el método de bisección en el intervalo cercano a la vecindad discontinua, este entregará un valor distinto a las raices de las funciones correspondientes.  

En este caso las raices son -2 y 2, y el método de bisección entrega como valor de convergencia 1. Esto al tomar el intervalo [0,1.5] , el cual cumple: $$p(a) \cdot p(b)< 0$$

Y es consistente con:

$$
\frac{|p_n - p_{n-1}|}{|p_n|} < \epsilon
$$
